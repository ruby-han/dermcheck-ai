```python
for path in path_full_three:
    img_id = os.path.splitext(os.path.basename(path))[0]
    background = Image.open(path)
    background = background.resize((224,224))
    background = background.convert("RGBA")
    overlay5 = overlay5.resize((224,224))
    overlay5 = overlay5.convert("RGBA")
    new_img = Image.blend(background, overlay5, 0.5) #the third param here is the level of transparency desired
    new_img = new_img.convert('RGB')
    new_img.save('./Data/overlayed_images_F_5/' + img_id +'.png')
```
